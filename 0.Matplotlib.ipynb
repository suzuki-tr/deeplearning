## グラフの表示

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,6,0.1) #0から6まで0.1刻みで生成
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1,linestyle="-",label='sin')
plt.plot(x,y2,linestyle="--",label='cos') #破線
plt.xlabel("x") #X軸ラベル
plt.ylabel("y") #Y軸ラベル
plt.title('sin & cos') #タイトル
plt.legend() #判例
plt.show()
```
